📢❗🚨 Just some quick reminders! 📢❗🚨

**DataValue**: The numeric value of the Chronic disease indcator so thefore, depending on the indcator, it can include
- a percentage

**Topic**= The overall category for chronic diseases or health bahvior
- obesity

- Diabetes

**Indicator** = An very specific measure inside the **Topic**
- percent of adults who smoke

**LocationDsec** = The state where the measure was taken


```python
import pandas as pd
import matplotlib.pyplot as plt
df= pd.read_csv("U.S._Chronic_Disease_Indicators.csv")
```

```python
#FILTER OUT EXTREME VALUES ONLY VALUES BETWEEN 0 AND 200 
df_rates=df[(df["DataValue"]>=0)&(df["DataValue"] <= 300)]
```

```python
# HISTOGRAM OF DataValue
df_rate_measures= df[df["DataValueType"].str.contains("Rate")]
df_rate_measures=df_rate_measures[df_rate_measures["DataValue"]<=300]
df_rate_measures["DataValue"].plot.hist(bins=30)
plt.title("Distribution of Chronic Disease Rates")
plt.xlabel("Rate (DataValue)")
plt.ylabel("Frequency")
```

```python
year_means=df_rates.groupby("YearStart")["DataValue"].mean().sort_index()
year_means.plot.line()
plt.title("Average Chronic Disease Indicator Value by Year (Filtered 0-300)")
plt.xlabel ("Year")
plt.ylabel("Average DataValue (%)")
```

```python
topic_means=df_rates.groupby("Topic")["DataValue"].mean().sort_values(ascending=False)
topic_means.plot.bar()
plt.title("Top 10 Topics by Average DataValue (Filtered 0-300)")
plt.xlabel ("Topic")
plt.ylabel("Average DataValue (%)")
```

# **Section 6: A Different Data Visualization**

```python
import pandas as pd
df=pd.read_csv("U.S._Chronic_Disease_Indicators.csv")
import matplotlib.pyplot as plt

df_year = df[(df["DataValue"] >= 0) & (df["DataValue"] <= 300)]
df_year.boxplot(column= "DataValue", by= "YearStart")
plt.title("Boxplot of DataValue by Year (0–300)")
plt.xlabel("Year")
plt.ylabel("DataValue")
```

✏️Summary of what is seen in the graphs✏️

I decided to create another visual representation with a box plot for DataValue based on YearStart with a filter applied (DataValue= 0-300) to provide an easier view of the distribution. I selected a box plot since it provides the ability to see how the median, spread, and presence of outlier values of chronic disease indicators vary over time. The box plots differ from the bar chart, histogram, and line graph used in section 3. The box plots reveal that the chronic disease indicators exhibit consistent right-skewness throughout all years with the majority of DataValue concentrated in a narrow range of 10-50. Additionally, each year has a significant number of extreme outliers (values ranging from 100-300), represented by the black vertical lines extending above year's box plot. These extreme values indicate that a small group of chronic disease indicators. Most likely these are associated with severe disease outcomes like hospitalization or moraltiy which account for the mahoirty of the extreme DataValues at teh higher end of the data value distributions. A second feature of the box plots revealed the median DataValue is decreasing from approximately 25-30 in 2025-2018 to lower levels in 2021-2022. As  previously demonstrated through the line graphs, this trend supports the notion of slight decreases in overall chronic disease rates in recent years. Ultimately, the box plot helps reinforce the idea that although the majority of chronic disease indcators remain at low values, there are substantial amount of high DataValue values present each year.
